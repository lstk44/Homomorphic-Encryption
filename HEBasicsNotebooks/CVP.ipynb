# Closest Vector Problem

## Imports

```python
import numpy as np
```

## Define Parameters



```python
GoodBasis = np.array([
    [1, 1],
    [1, -1]
])

BadBasis = np.array([
    [3, 2],
    [4, 3]
])

Secret = np.array([3, 1])
SecretWithError = np.array([3.4, 0.6])
```

## CVP

```python
# Find vector combination of basis that is closest to point
def cvp(basis, point):
    # Find the coefficients of the vector combination
    coefficients = np.linalg.solve(basis, point)
    # Round coefficients to nearest integer
    coefficients = np.round(coefficients).astype(int)
    # Calculate the vector combination
    combination = list(np.round(np.dot(basis, coefficients)).astype(int))
    print(f'Vector combination: vector1 * {coefficients[0]} + vector2 * {coefficients[1]} = {combination}')
```

## Without Error

```python
cvp(GoodBasis, Secret)
```

```python
cvp(BadBasis, Secret)
```

## With Error

```python
cvp(GoodBasis, SecretWithError)
```

```python
cvp(BadBasis, SecretWithError)
```
